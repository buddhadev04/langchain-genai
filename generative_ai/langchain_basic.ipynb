# Groq Model Intregation


### get response



```python
from langchain.chat_models import init_chat_model
import os
from dotenv import load_dotenv
load_dotenv()
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")  
model = init_chat_model("groq:openai/gpt-oss-120b")
response = model.invoke("Tell me a joke")
response.content
```

### Stream using the same model

```python
for chunk in model.stream("Write a short story about a robot learning to paint"):
    print(chunk.text, end="", flush=True)
```

### Batch requests -- it's used to parallealy send requests which llm process parallely

```python
model.batch(["What is the capital of France?", "What is the largest mammal?", "How do air planes fly?"], config={"max_concurrency": 5})
```
