## Tools


### LLMs, that create new, original content—including text, images, audio, video, and code—based on user-provided prompts. to get rid of from halusination


```python
import os
from langchain.chat_models import init_chat_model
from dotenv import load_dotenv
load_dotenv()
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")  
model = init_chat_model("groq:openai/gpt-oss-120b")
response = model.invoke("write me an eassy on AI")
response
```

### tools

```python
from langchain.tools import tool

@tool()
def get_current_weather(location: str) -> str:
    """Get the current weather in a given location""" # This is a dockstring that describes the function and its parameters
    return f"The current weather in {location} is sunny with a high of 25 degrees Celsius."

model_with_tools = model.bind_tools([get_current_weather])
response = model_with_tools.invoke("What is the current weather in Gurgaon?")
response.tool_calls # this will show the tool calls that were made by the model
```

### tool execution loop

```python
# model generates tool call
messages = [{'role': 'user', 'content': 'What is the current weather in Gurgaon?'}]
response = model_with_tools.invoke(messages)
messages.append(response)

# execute tool call and collect tool response
for tool_call in response.tool_calls:
    tool_result = get_current_weather.invoke(tool_call)
    messages.append(tool_result)

# pass the result back to the model
final_response = model_with_tools.invoke(messages)
final_response.text
```
