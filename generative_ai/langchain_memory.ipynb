```python
import os
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver

agent = create_agent(
    model="groq:openai/gpt-oss-120b",
    checkpointer=InMemorySaver()
)
```

```python
# run with thread id
config = {"configurable": {"thread_id": "my_thread_id"}}
```

```python
from langchain_core.messages import HumanMessage
result = agent.invoke({
    "messages": [HumanMessage(content="hey, my name is mono")],
},config=config)
result
```

```python
result = agent.invoke({
    "messages": [HumanMessage(content="hey, what is my name?")],
},config=config)
result
```

```python
# use other config
config = {"configurable": {"thread_id": "another_thread_id"}}
result = agent.invoke({
    "messages": [HumanMessage(content="hey, what is my name?")],
},config=config)
result
```

```python
result = agent.invoke({
    "messages": [HumanMessage(content="my name is mono")],
},config=config)
result
```

```python
result = agent.invoke({
    "messages": [HumanMessage(content="what i asked you before?")],
},config=config)
result
```
