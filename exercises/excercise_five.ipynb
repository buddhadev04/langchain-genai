### Exercise 5: The Reasoning Enforcer

Topic: Structured Output & Prompt Engineering


Goal: Force the model to "Think" before it "Answers."

```python
import os
from dotenv import load_dotenv
load_dotenv()
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from pydantic import BaseModel
from langchain.agents import create_agent
class ChainOfThought(BaseModel):
    """Chain of thought."""
    logic_steps: list[str]
    final_answer: str

agent = create_agent(
    model="groq:openai/gpt-oss-120b",
    tools=[],
    response_format=ChainOfThought,
)
prompt = """
You must solve the user's problem.

You are strictly forbidden from providing the final_answer
until you have listed at least 3 logical steps in logic_steps.

Problem:
If I have 2 apples and eat 1, then buy 5 more, how many do I have?
"""
response = agent.invoke({
    'messages': [{"role": "user", "content": prompt}]
})
result = response['structured_response']
print(result)
print("Logic steps:")
for step in result.logic_steps:
    print(f"- {step}")
print(f"Final answer: {result.final_answer}")

```
