### Exercise 4: Parallel Tool Invocation

Topic: Agents & Efficiency


Goal: Prove the agent can handle multiple tasks in one turn.

```python
import os
from dotenv import load_dotenv
load_dotenv()
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from langchain.agents import create_agent
from langchain_core.tools import tool

@tool
def get_weather(city: str):
    """Get current weather."""
    return f"22 degrees in {city}"

@tool
def get_news(city: str):
    """Get latest news."""
    return f"Local festival starts today in {city}."


agent = create_agent(
    model="groq:openai/gpt-oss-120b",
    tools=[get_weather, get_news],
)

prompt = """
Call both tools get_weather and get_news for Paris in a single step.
Return both results together.
Do not call them sequentially.
"""

response = agent.invoke({
    "messages": [{"role": "user", "content": prompt}]
})
print(response)
# 🔍 Inspect tool calls
if hasattr(response, "tool_calls"):
    print("Tool Calls:")
    for call in response.tool_calls:
        print(call["name"])
```
